```python
# Import Statements

from xml.dom import minidom
import pandas
import datetime as dt
from matplotlib import pyplot as plt
```

```python
# Parse the biking data as a Document  
biking_data = minidom.parse("biking3-15-2012.tcx.xml")

# Make list of trackpoints
trackpoints = biking_data.getElementsByTagName("Trackpoint")
```

```python
# Get time and altitude data into a list

time = []
altitude = []

for idx in range(804):
    time.append(trackpoints[idx].getElementsByTagName("Time")[0].firstChild.data)
    altitude.append(trackpoints[idx].getElementsByTagName("AltitudeMeters")[0].firstChild.data)
    
```

```python
# Convert time from string to datetime.datetime object 

time_fix = []

for t in time:
    time_fix.append(dt.datetime.strptime(t, '%Y-%m-%dT%H:%M:%SZ'))
```

```python
# View time_fix data
time_fix[0:5]
```

```python
# Convert alt strings to float

altitude_fix = []

for alt in altitude:
    altitude_fix.append(float(alt))
```

```python
# View altitude_fix data
altitude_fix[0:5]
```

```python
plt.plot(time_fix, altitude_fix)
plt.xlabel("time")
plt.ylabel("altitude (m)")
plt.title("Altitude vs. Time")

```

### Summary

In the chart above, we observe bimodal plot of the altitude vs time. As the cyclist rides his bike, he/she climbs a height of 330m around with a total ascension of 80m.
